In the very beginning, I tried cleaning the dataframe by commenting out the columns that I would not need at the time, instead of dropping. Then,
I converted my last_review into datetime from its original data type. I also I decided to remove the all the NaN values, so it would be easier to read
the available data. 

```python
import pandas as pd
df = pd.read_csv(r"C:\Users\theal\Downloads\AB_NYC_2019 (1).csv")
pd.set_option('display.max.columns', 300)
pd.set_option('display.max.rows', 50)
df1 = df[[#'id', 
    'name', 'host_id', 'host_name', 
    'neighbourhood_group','neighbourhood', 
    #'latitude', 'longitude', 
     'room_type', 'price',
       'minimum_nights', 'number_of_reviews', 'last_review',
       'reviews_per_month', 'calculated_host_listings_count',
       'availability_365']]
df1['last_review'] = pd.to_datetime(df1.last_review)
#df1.dtypes
df1 = df1.dropna()
df1.isna().sum()
df1.dtypes

```

#1. Which hosts are the busiest and why?
I would just compare the amount of host listings and the host columns. I can infer if a host has more properties, the probability of the
host being busy based on the amount of properties, is likely high.

```python
#df1[['calculated_host_listings_count','number_of_reviews']].max()
df1_sorted = df1.sort_values(by=['host_name','calculated_host_listings_count'],ascending=False)
df1_sorted
```

#2. How many neighborhood groups are available and which shows up the most?
I would just request all the neighborhood groups in a value count to see how many places there are in each neighborhood.

```python
neighbourhood_counts = df1['neighbourhood_group'].value_counts()
neighbourhood_counts
```

#3. Are private rooms the most popular in manhattan?
For this question I grouped by neighborhood_group to see the room's type total count. There I could see which type of room is
available in which group. 

```python
housing_popularity_manhattan = df1.groupby('neighbourhood_group')['room_type'].value_counts()
housing_popularity_manhattan
#df1.dtypes
```

#4. Which hosts are the busiest and based on their reviews?
This one was a bit more challenging because I wanted to compare the most relevant values, which in my case were number of reviews and
reviews per month by host. The reason being if the customer had many reviews in a month, I can infer that the host has a quick turnaround.
I can compare mean by number of reviews to get an idea of average of reviews per month.
Then, I want to see how recent is the most recent review, to get a general idea of perhaps activity amongst host.

```python
#4. Which hosts are the busiest and based on their reviews?
df1.dtypes
busiest_host = df1.groupby('host_name').agg(last_review=('last_review',max), reviews_per_month=('reviews_per_month', 'mean'),number_of_reviews=('number_of_reviews','mean'))
sorted_host = busiest_host.sort_values(by=['last_review', 'reviews_per_month','number_of_reviews'], ascending=False)
sorted_host
```

#5. Which neighorhood group has the highest average price?
I grouped the neighbourhood column with price column and just retreived the mean from price. By comparing them side by side, I can
see which neighbourhood has the average highest price, and sort the columns by descending order.

```python
#5. Which neighorhood group has the highest average price?
df1.dtypes
housing_average = df1.groupby('neighbourhood_group').agg(price=('price', 'mean'))
high_price_average = housing_average.sort_values(by=['price'], ascending=False)
high_price_average.head(15)
```

#6. Which neighborhood group has the highest total price? In this cell, I just grouped by neighbourhood and added the max value of price. After,
I just sorted from high to low.

```python
#6. Which neighborhood group has the highest total price?
housing_highest_cost = df1.groupby('neighbourhood_group').agg(price=('price', 'max'))
high_price_cost = housing_highest_cost.sort_values(by=['price'], ascending=False)
high_price_cost.head(15)
```

#7. Which top 5 hosts have the highest total price? price + host 
Here I grouped by host and added the max value of the price column. Then I sorted by descending value. By adding head to 15, I can see
the top 15 rows based on the grouping in a descending order.  

```python
#7. Which top 5 hosts have the highest total price? price + host 
#df1.dtypes
highest_price_per_host = df1.groupby('host_name').agg(price=('price', 'max'))
high_price_host_sorted = highest_price_per_host.sort_values(by=['price'], ascending=False)
high_price_host_sorted.head(15)
```

```python
#8. Who currently has no (zero) availability with a review count of 100 or more?
df1.dtypes
# host_2_review_avail = df1.groupby('host_name').agg(availability_365=('availability_365' == 0),number_of_reviews=('number_of_reviews' > 100))
# host_2_review_avail
no_year_round_avail = (df1['availability_365'] == 0).sum()
no_year_round_avail
```

#9. What host has the highest total of prices and where are they located?
Here I grouped a few columns to compare, which are host name, neighbourhood and price. I set price to highest value based on price,
to see, based on the highest price. TThere I can see which host and where. 

```python
highest_location_cost_per_host = df1.groupby(['host_name', 'neighbourhood'])['price'].agg(['max','sum'])
#max_host, max_location = highest_location_cost_per_host.idxmax()
tab_highest_cost_per_host_df = highest_location_cost_per_host
tab_highest_cost_per_host_df.sort_values(by='max', ascending=False)
```

#10. When did Danielle from Queens last receive a review?
Here, the approach was to group host and neighborhood group with last review. To see when Danielle received the last review in queens. Again, I used
the max method to get the most recent or highest value from the last review column.

```python
#df1.dtypes
host_danielle = df1.groupby(['host_name','neighbourhood_group']).agg('last_review').max()
# host_danielle_sorted = host_danielle.sort_values('last_review', ascending=False)
host_danielle = host_danielle.loc['Danielle']
host_danielle
```

#11. Which host has the most listings?
I grouped by host and calculated host listings count. My thoughts were at first to group the columns and sort by aggregating max, but it did not make
full sense. So i added sum because the sum is the total values per host, or at least that is how I am interpreting it. Sum shows the total per host.

```python
hostes_with_the_mostest = df1.groupby('host_name')['calculated_host_listings_count'].agg(['max','sum'])
hostes_with_the_mostest.sort_values(by='max',ascending=False)
```

#12. How many listings have completely open availability? For this question, I created a conditional statement based on the availability
column and the integer 365. It returns a boolean, so I just added all them ones. There I received a full count of all the completely open 
availibility throughout the year. 

```python
#df1.dtypes
year_round_avail = (df1['availability_365'] == 365).sum()
year_round_avail
```

#13. What room types have the highest review numbers?
The room types with the maximum reviews is what I grouped together. Each room type displays the highest number of reviews. Lastly
sort the columns in descending order.

```python
#df1.dtypes
high_value_room = df1.groupby('room_type')['number_of_reviews'].max()
high_value_room.sort_values(ascending=False)
```
